```python
import pandas as pd
import requests
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# # Load RNA-seq data
# rna_seq_url = 'https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE52462&format=file&file=GSE52462%5FRNA%5Fseq%5Fmatrix.txt.gz'
# rna_seq_df = pd.read_csv(requests.get(rna_seq_url).text, sep='\t')

# # Load single cell data
# single_cell_url = 'https://github.com/MarioniLab/MouseRNAseqScreens/raw/master/2917_BrainMyeloid/data/allen_mouse_brain_atlas_expression_matrix_filtered.csv'
# single_cell_df = pd.read_csv(requests.get(single_cell_url).text)


# Simulate RNA-seq dataset
rna_seq_df = pd.DataFrame(np.random.randint(0,100,size=(100, 20000)), columns=range(20000))

# Simulate single-cell dataset
single_cell_df = pd.DataFrame(np.random.normal(0, 1, size=(5000, 100)), columns=range(100))


# Standardize and PCA
scaler = StandardScaler()
rna_seq_scaled = scaler.fit_transform(rna_seq_df)
single_cell_scaled = scaler.fit_transform(single_cell_df)

pca = PCA(n_components=2)
rna_seq_pca = pca.fit_transform(rna_seq_scaled)
single_cell_pca = pca.fit_transform(single_cell_scaled)

# Visualize PCA plots
plt.scatter(rna_seq_pca[:,0], rna_seq_pca[:,1], alpha=0.5, label='RNA-Seq')
plt.scatter(single_cell_pca[:,0], single_cell_pca[:,1], alpha=0.5, label='Single-Cell')
plt.legend()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Plot Comparison')
plt.show()

print("Single-cell data shows finer differentiation of cell subpopulations")
```
